# Contrôle du vna (ZND Rohde&Schwarz)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pyvisa
rm = pyvisa.ResourceManager()
znd=rm.open_resource('znd')
```

```python
# Check connection
znd.query('*IDN?')
```

## Récupération de la trace active

```python
# Span
channel = 1
start = float(znd.query(f"SENS{channel}:FREQ:STAR?"))
stop = float(znd.query(f"SENS{channel}:FREQ:STOP?"))
npoints = int(znd.query(f"SENS{channel}:SWE:POIN?"))
freq = np.linspace(start,stop,npoints)
```

```python
# Format binaire
znd.write("FORM REAL ,32")
```

```python
# Trace active
data = znd.query_binary_values(f'CALC{channel}:DATA? SDAT',container=np.ndarray)
Sij = data.view(np.complex64)
```

```python
d=5e-2
plt.plot(freq,np.unwrap(np.angle(Sij)))
p=np.polyfit(freq,np.unwrap(np.angle(Sij)),1)
plt.plot(freq,np.polyval(p,freq))
#plt.plot(freq,freq*2*3.14*d/(3*10**8))
```

```python
p
2*np.pi*d/(p[0]*1e8)
```

```python
# Save data
import h5py
with h5py.File("Vitesse_c.hdf5", "w") as f:
    f.create_dataset("freq", data=freq)
    f.create_dataset("phi", data=np.unwrap(np.angle(Sij)))
```
