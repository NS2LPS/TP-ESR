# Sauver des données au format HDF5

Le format HDF5 (Hierarchical Data Format version 5) est un format de fichier binaire conçu pour stocker, organiser et manipuler de grandes quantités de données de manière efficace, portable et structurée. Il est très utilisé en science, ingénierie, intelligence artificielle et calcul haute performance.

HDF5 repose sur une structure hiérarchique, comparable à un système de fichiers. Les données peuvent être regroupées par groupes et on peut ajouter des attributs aux données pour préciser certains paramètres.

Le format HDF5 peut être lu par tous les logiciels de traitement de donnée ou de calcul (Matlab, Python, Julia, R, ...).

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
import h5py
```

```python
t = np.linspace(0,1,1024)
y = np.random.randn(1024)
plt.plot(t,y)
```

## Sauver des données

```python
# Save data
with h5py.File("mytestfile.hdf5", "w") as f:
    f.create_dataset("dataset000/t", data=t)
    f.create_dataset("dataset000/y", data=y)
    attrs = f['dataset000'].attrs
    attrs.update(dict(time=time.time()))
```

## Charger des données

```python
# Read data
with h5py.File("mytestfile.hdf5", "r") as f:
    i = 0
    ds = f[f'dataset{i:03d}']
    t = np.array(ds['t'])
    y = np.array(ds['y'])
    print('Time',ds.attrs['time'])
```

```python
plt.plot(t,y)
```
